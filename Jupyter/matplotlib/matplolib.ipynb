```python
# area graph
days     = [1, 2, 3, 4, 5]

sleeping = [7, 8, 6, 11, 7]
eating   = [2, 3, 4, 3, 2]
working  = [7, 8, 7, 2, 2]
playing  = [8, 5, 7, 8, 13]

plt.plot([], [], color='m', label="sleeping", linewidth=5)
plt.plot([], [], color='c', label="eating",   linewidth=5)
plt.plot([], [], color='r', label="working",  linewidth=5)
plt.plot([], [], color='k', label="playing",  linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing,
              colors=['m', 'c', 'r', 'k'])

plt.title('area graph')
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.show()

```

```python
# pi chart
import matplotlib.pyplot as plt
import matplotlib.style
slices   = [7, 2, 2, 13]

#sleeping = [7, 8, 6, 11, 7]
#eating   = [2, 3, 4, 3, 2]
#working  = [7, 8, 7, 2, 2]
#playing  = [8, 5, 7, 8, 13]

sleeping = "sleeping"
eating = 'eating'
working = 'working'
playing = 'memes'

activities =[sleeping, eating, working, playing]
cols = ['m', 'c', 'r', 'y']
plt.pie(slices, 
       labels=activities, 
       colors=cols,
       startangle=90, 
       shadow=True,
       explode=(0, 0, 0.1, 0),
       autopct='%1.1f%%')

plt.title('Pie Plot')
plt.show()
```
