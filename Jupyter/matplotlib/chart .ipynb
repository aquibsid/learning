```python
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
```

```python
# scatter plot 
x = np.arange(0, 8, 1) # [0, 1, 2, 3, 4, 5, 6, 7]
y = np.random.randint(8, size=8)
#y = [4, 2, 3, 5, 6, 6, 1, 8]
#print(y)
 
plt.scatter(x, y, label='skitscat',marker="v", color='k', s=30)

plt.xlabel('X')
plt.ylabel('y')
plt.title("scatter plot")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# stack plot

days = np.arange(0, 5, 1)

sleeping = [1, 2, 4, 3, 6]
eating   = [3, 5, 7, 2, 3]
working  = [5, 8, 1, 3, 2]
playing  = [5, 3, 4, 1, 2]


plt.plot([], [], label='sleeping', color='r', linewidth=5)
plt.plot([], [], label='eating',   color='b', linewidth=5)
plt.plot([], [], label='working',  color='g', linewidth=5)
plt.plot([], [], label='playing',  color='y', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing,
              colors=['r', 'b', 'g', 'y'])

plt.xlabel('X')
plt.ylabel('y')
plt.title("scatter plot")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python

# pie chart


sleeping = [1, 2, 4, 3, 6]
eating   = [3, 5, 7, 2, 3]
working  = [5, 8, 1, 3, 2]
playing  = [5, 3, 4, 1, 2]


slices = [3, 5, 1, 7]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['r', 'b', 'g', 'y']

plt.pie(slices,
        labels=activities,
        colors = cols,
        startangle = 90,
        explode = (0, 0, 0.1, 0),
        autopct = '%1.1f%%')


plt.title("Pie chart")
#plt.legend()
plt.show()




```
