```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 5)) # figsize=(width, length), it is relative size to fig
#df.plot(ax=axes[1])
fig.subplots_adjust(hspace=0.7)  # space b/w two graphs

df['key'].plot(ax = axes[0])
df['lock'].plot(ax = axes[1])

axes[0].set_title('graph 0', fontsize=15)
axes[0].set_xlabel('x-axis', fontsize=15)
axes[0].set_ylabel('y-axis', fontsize=15)
axes[0].legend(['line'])

axes[1].set_title('graph 1', fontsize=15)
axes[1].set_xlabel('x-axis', fontsize=15)
axes[1].set_ylabel('y-axis', fontsize=15)
axes[1].legend(['line'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x = np.linspace(0, 3, 20) # x-axis ploting 20 points b/w 0 and 3
ax.plot(x, x,    label='linear',    marker='o',markerfacecolor='k', color='b', linestyle='--', linewidth=0.5)
ax.plot(x, x**2, label='quadratic', marker='>', color='g', linewidth=0.5)
ax.plot(x, x**3, label='cubic',     marker='*', color='y', linewidth=0.5)

ax.legend()
ax.set_title("Graph")
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.plot()
```

```python
import numpy as np
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 5))
fig.subplots_adjust(hspace=0.8)# space b/w graph

#fig, ax = plt.subplots()
x = np.linspace(0, 3, 20)
ax[0][0].plot(x, x, label='linear', marker='o',color='b', linewidth=0.5)
ax[0][0].set_title("linear ")
ax[0][0].set_xlabel('x-axis')
ax[0][0].set_ylabel('y-axis')

ax[0][1].plot(x, x**2, label='quadratic', marker='>', color='g')
ax[0][1].set_title("Quadratic")
ax[0][1].set_xlabel('x-axis')
ax[0][1].set_ylabel('y-axis')

ax[1][0].plot(x, x**3, label='cubic', marker='^', color='y')
ax[1][0].set_title("Cubic")
ax[1][0].set_xlabel('x-axis')
ax[1][0].set_ylabel('y-axis')

ax[1][1].plot(x, x, label='linear', marker='o',color='b', linewidth=0.5)
ax[1][1].plot(x, x**2, label='quadratic', marker='>', color='g')
ax[1][1].plot(x, x**3, label='cubic', marker='^', color='y')

ax[1][1].set_title("ALL")
ax[1][1].set_xlabel('x-axis')
ax[1][1].set_ylabel('y-axis')


#ax.legend()

plt.show()
```
